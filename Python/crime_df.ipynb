```python
import pandas as pd
import numpy as np
from scipy.stats import chi2_contingency
from sklearn.preprocessing import LabelEncoder
import seaborn as sns
import matplotlib.pyplot as plt

def cramers_v(x, y):
    confusion_matrix = pd.crosstab(x, y)
    chi2 = chi2_contingency(confusion_matrix)[0]
    n = confusion_matrix.sum().sum()
    phi2 = chi2 / n
    r, k = confusion_matrix.shape
    return np.sqrt(phi2 / min(k - 1, r - 1)) if min(k - 1, r - 1) > 0 else np.nan


def is_categorical(series, threshold=20):
    return series.dtype == 'object' or series.nunique() <= threshold

def mixed_correlation_matrix(df):
    cols = df.columns
    matrix = pd.DataFrame(index=cols, columns=cols)

    for col1 in cols:
        for col2 in cols:
            x, y = df[col1], df[col2]

            if is_categorical(x) and is_categorical(y):
                value = cramers_v(x, y)
            elif np.issubdtype(x.dtype, np.number) and np.issubdtype(y.dtype, np.number):
                value = x.corr(y)
            else:
                if is_categorical(x):
                    x = LabelEncoder().fit_transform(x.astype(str))
                if is_categorical(y):
                    y = LabelEncoder().fit_transform(y.astype(str))
                value = pd.Series(x).corr(pd.Series(y))

            matrix.loc[col1, col2] = round(value, 3)

    return matrix.astype(float)


correlation_matrix = mixed_correlation_matrix(crime)


plt.figure(figsize=(16, 12))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm', center=0)
plt.title("Correlation Matrix (Categorical + Numeric)")
plt.xticks(rotation=45)
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
corr_matrix = crime.corr(numeric_only=True)


mask = np.triu(np.ones_like(corr_matrix, dtype=bool))
filtered_corr = corr_matrix.mask(mask)


corr_pairs = (
    filtered_corr.stack()
    .reset_index()
    .rename(columns={'level_0': 'Feature 1', 'level_1': 'Feature 2', 0: 'Correlation'})
)

corr_pairs['AbsCorrelation'] = corr_pairs['Correlation'].abs()
corr_pairs = corr_pairs.sort_values(by='AbsCorrelation', ascending=False).drop(columns='AbsCorrelation')
print(corr_pairs.head(20))  

```

```python

start_col = 'Crime Type'
cols_from_crime_type = crime.columns[crime.columns.get_loc(start_col):]

crime_encoded = crime.copy()
for col in cols_from_crime_type:
    if crime_encoded[col].dtype == 'object':
        crime_encoded[col] = pd.factorize(crime_encoded[col])[0]

corr = crime_encoded[cols_from_crime_type].corr()

mask = np.triu(np.ones_like(corr, dtype=bool))

plt.figure(figsize=(12, 10))
sns.heatmap(corr, mask=mask, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title("Correlation Heatmap (from 'Crime Type' onwards)", fontsize=14)
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```
